```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
data = []
base_url = "https://tiki.vn/"
input = 'pedal piano'
```

```python
# Trả về info của data bằng try except
def get_info(item,css,other):
    try:
        x = item.find_element(By.CSS_SELECTOR,css).text
    except:
        x = other
    return x
```

```python
def get_website(url = base_url):
    cnt = 1
    driver = webdriver.Chrome()
    driver.get(url)
    
    time.sleep(5) # Chờ pop up quảng cáo
    try:
        close_btn = driver.find_element(By.CSS_SELECTOR,"img[alt='close-icon']")
        close_btn.click()
    except:
        print("Không thấy pop up quảng cáo")
    
    # Bắt đầu tìm kiếm
    search_bar = driver.find_element(By.CSS_SELECTOR,'.sc-dec0a11d-2')
    search_bar.send_keys(input)
    find_btn = driver.find_element(By.CSS_SELECTOR,"button[data-view-id='main_search_form_button']")
    find_btn.click()

    while url:
        # Thu thập dữ liệu các sản phẩm
        time.sleep(3)
        product = driver.find_elements(By.CSS_SELECTOR,"div[class='info']")
        for item in product:
            price = get_info(item,"div[class='price-discount__price']","0₫")
            describe = get_info(item,"h3[class='sc-68e86366-8 dDeapS']","Không có mô tả")
            quantity = get_info(item,"span.quantity","Đã bán 0")
            data.append([price,describe,quantity])
            print(f"Item: {cnt}\nPrice: {price}\nDescribe: {describe}\nQuantity: {quantity}\n\n")
            cnt+=1

        # Thực hiện sang trang tiếp theo
        try:
            next_btn = driver.find_element(By.CSS_SELECTOR,"a.arrow:not(.disabled) img[alt='arrow-right']")
            next_btn.click()
        except:
            break
```

```python
get_website()
```

```python
df = pd.DataFrame(data,columns=['price','description','quantity'])
df.to_csv('tiki_product.csv',index=False,encoding='utf-8-sig')
print('Saved!')
```
