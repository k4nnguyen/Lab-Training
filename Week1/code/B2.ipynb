## requests + BeautifulSoup (Thực hành)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = url = 'https://books.toscrape.com/'
cnt = 1
data = []
while url:
    print(cnt)
    changed = 0 # Nếu không có phần next thì break ra
    rq = requests.get(url)
    soup = BeautifulSoup(rq.content)
    for block in soup.findAll('article', class_='product_pod'):
        book_name = block.find('a').contents[0]['alt']
        book_price = block.find('p',class_='price_color').contents[0]
        data.append([book_name,book_price])
        for block in soup.findAll('a'):
            if(block.contents[0] == 'next'):
                changed = 1
                if(cnt > 1):
                    url = base_url + "/catalogue/" + block['href']
                else:
                    url = base_url + "/" + block['href']
                print(url)
    cnt+=1
    if(changed == 0): break
df = pd.DataFrame(data)
df.to_csv('book.csv',index=False, encoding='utf-8-sig')
```
